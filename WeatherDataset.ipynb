### Create a visual that shows disaster counts per year

```python
# Import 
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
path = r"C:\Users\jaded\Desktop"
os.chdir(path)
sourcefiles = os.listdir(path)
df = pd.read_csv("Weather Data.csv",sep = ",")
df.head()
```

### Create a histogram to show the distribution of temperatures

```python
df['Season'] = df['Season'].astype('category')

plt.figure(figsize = (10,6))
sns.histplot(df, x = 'Temp')
plt.title("Histogram of Tempratures")
plt.show()
```

```python
# the data is distributed in where it shows a normal distribution
# the majority temperature is 60
```

```python
boxplot = px.box(df, x = "Season", color = "Location", y = "Temp",
                 color_discrete_sequence = px.colors.cyclical.Edge)

# Update titles (main and axes) and format
boxplot.update_layout(title = {"text": "Seaonsal Temperature Distribution",
                               "xanchor": "left", "yanchor": "top", "x": 0.1}, 
                               legend_title_text = "Location", xaxis_title = "Season", yaxis_title = "Temperature (F)")
boxplot.show()
```
