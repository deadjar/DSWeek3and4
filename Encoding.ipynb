# Data Encoding - Disaster Dataset

### Encode 3 different categorical columns in this dataset using scikit-learn techniques, OneHotEncoding and OrdinalEncoding 

```python
import os 
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
path = r"C:\Users\jaded\Desktop"
os.chdir(path)
sourcefiles = os.listdir(path)
df = pd.read_excel("public_emdat_project.xlsx")
```

```python
df_clone = df.copy()
df_clone2 = df.copy()
```

#### Ordinal Encoding

```python
ord_encoder2 = OrdinalEncoder()
df_clone2['Region'] = ord_encoder2.fit_transform(df_clone2[['Disaster Type']])
df_clone2[['Disaster Type','Region']].head(5)
```

#### Ordinal Encoding

```python
ord_encoder = OrdinalEncoder()
df_clone["Disaster Type Code"] = ord_encoder.fit_transform(df_clone[["Disaster Type"]])
df_clone[["Disaster Type", "Disaster Type Code"]].head(10)
```

#### One Hot Encoding

```python
onehot_encoder_clone = df.copy()
onehot_encoder = OneHotEncoder()
onehot_results = onehot_encoder.fit_transform(onehot_encoder_clone[['Disaster Type']])
onehot_df = pd.DataFrame(onehot_results.toarray(), columns = onehot_encoder_clone["Disaster Type"].unique())

onehot_encoder_clone = onehot_encoder_clone.join(onehot_df)
df.head(10)
```
